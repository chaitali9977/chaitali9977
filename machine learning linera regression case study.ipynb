```python
#importing necessary packages
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance.csv")
```

```python
data.head()
```

```python
data.info() #basic idea that we get from the data
```

```python
data.describe()#statistical analysis
```

```python
#data preprocessing
#convert categorical variables into numerical variables using one hot encoding technique
```

```python
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
```

```python
data.head()
```

```python
data['sex']=data['sex'].apply({'female':0,'male':1}.get)
```

```python
data.head()
```

```python
data['region']=data['region'].apply({'southeast':0,'northeast':2,'northwest':3,'southwest':1}.get)
```

```python
data.head()
```

```python
#third step in machine learning is dividing the data
#dividing the data in dependent and independent columns
```

```python
X=data[['age','sex','bmi','children','smoker','region']]#independent features from the dataset

y=data['charges'] #dependent column from the dataset
```

```python
X.head()
```

```python
y.head()
```

```python
# 4 th step of machine learning  to splitting the dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)#20%of the dataset will be used for testing(evaluation)
#80% of the data is to be used for training purposes
```

```python
#create the machine learning model
from sklearn.linear_model import LinearRegression
linearmodel=LinearRegression()
linearmodel.fit(X_train,y_train) # training is taking place
```

```python
#predictions
linearmodel.predict(X_test)
```

```python
#calculating the accuracy of the model
linearmodel.score(X,y) # evaluation is taking polace
```

```python
#predictions with respect to new dataset
data_new={'age':32,'sex':1,'bmi':25,'children':1,'smoker':0,'region':3}
```

```python
index=[1]# serial number
my_data=pd.DataFrame(data_new,index)
```

```python
print(my_data)
```

```python
#charges allocated
my_data_charges=linearmodel.predict(my_data)
print('the charges allocated to the new customer is', my_data_charges)
```
